## Importación de las librerías necesarias

```python
%matplotlib inline
import numpy as np
from scipy import misc
import imageio as img
import matplotlib.pyplot as plt
```

### Importación de la imagen capturada por el satélite, y almacenada en un ndarray

```python
from skimage import data

photo_data = img.imread('./wifire/sd-3layers.jpg')
type(photo_data)
```

### Mostrar la imagen con pylot

```python
plt.figure(figsize=(15, 15))
plt.imshow(photo_data)
```

```python
photo_data.shape
```

La forma (shape) del ndarray indica que se trata de una matriz de tres capas. El primer número (3.275) indica la altura de la imagen, el segundo (4.797) indica la achura, y el tercero (3) indica las tres capas que forman un color, es decir, RGB.

<h1 style="color: gray;"><u>Mapeado del color en la imagen</u></h1>

Los colores de cada uno de los píxeles de la imagen indican una información:
<ul>
<li><p style="font-size: 22px; color: red;"><b>ROJO</b> indica la altitud.</p></li>
<li><p style="font-size: 22px; color: blue;"><b>AZUL</b> indica la exposición.</p></li>
<li><p style="font-size: 22px; color: green;"><b>VERDE</b> indica la inclinación.</p></li>
</ul>

```python
# número de elementos del ndarray
photo_data.size
```

```python
# valor mínimo y máximo del ndarray
print("Valor mínimo: " + str(photo_data.min()))
print("Valor máximo: " + str(photo_data.max()))
```

```python
# media del valor de los píxeles
photo_data.mean()
```
